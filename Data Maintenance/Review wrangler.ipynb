# This is the code in analysing the star rating results for each charging station

```
#all libraries used
import pandas as pd
```

```
#update the path where the reviews data is saved
rev_path = "/content/drive/My Drive/B3_FIT5120FIT5122/Iteration 2/Dataset/Python Code/reviews.csv"
rev_df = pd.read_csv(rev_path, delimiter=";")
rev_df

#prepare data
#get average rating summary
ave_rating = rev_df.groupby('station_id')['rating'].mean().reset_index().sort_values(by='rating', ascending=True).set_index('station_id')
#get count of stations that are tagged as not working
not_working = rev_df.groupby('station_id')['is_working'].sum().reset_index().sort_values(by='is_working',ascending=False).set_index('station_id')

#https://robertmitchellv.com/blog-bar-chart-annotations-pandas-mpl.html
#graph the data
ax1 = ave_rating.plot(kind='barh', figsize=(10,7),color="yellow", fontsize=13);
ax1.set_alpha(0.8)
ax1.set_title("Average Rating Chart", fontsize=18)
ax1.set_ylabel("ID", fontsize=12)
ax1.set_xlabel("Average Rating", fontsize=12);
# invert for largest on top 
ax1.invert_yaxis()

ax2 = not_working.plot(kind='barh', figsize=(10,7),color="coral", fontsize=13);
ax2.set_alpha(0.8)
ax2.set_title("Not Working Chart", fontsize=18)
ax2.set_ylabel("ID", fontsize=12)
ax2.set_xlabel("Total who said the charger is not working", fontsize=12);

# invert for largest on top 
ax2.invert_yaxis()


#create new files for additional analysis of the user
ave_rating.to_csv(r'ave_rating.csv', index = False)
not_working.to_csv(r'ave_rating.csv', index = False)
```
